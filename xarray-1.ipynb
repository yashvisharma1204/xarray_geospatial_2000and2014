```python
import xarray as xr
import matplotlib.pyplot as plt
```

# Load datasets

```python
data1 = xr.open_dataset('GLDAS_CLSM025_D.A20000101.020.nc4')
data2 = xr.open_dataset('GLDAS_CLSM025_D.A20141230.020.nc4')
```

```python
data1
```

```python
data2
```

## Section 1: Mean Surface Temperature Comparison
This section compares the mean surface temperature for the years 2000 and 2014.

```python
surface_temp1 = data1['AvgSurfT_tavg']
surface_temp2 = data2['AvgSurfT_tavg']
```

```python
mean_temp1 = surface_temp1.mean(dim='time')
mean_temp2 = surface_temp2.mean(dim='time')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
mean_temp1.plot(ax=axes[0])
axes[0].set_title('Mean Surface Temperature in 2000')
mean_temp2.plot(ax=axes[1])
axes[1].set_title('Mean Surface Temperature in 2014')
plt.tight_layout()
plt.show()
```

## Section 2: Mean Rainfall, Snowfall, Soil Moisture, and Air Temperature
This section compares the mean rainfall, snowfall, soil moisture, and air temperature for the years 2000 and 2014.

```python
mean1 = rain1.mean(dim='time')
mean2 = rain2.mean(dim='time')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
mean1.plot(ax=axes[0])
axes[0].set_title('Mean Rainfall in 2000')
mean2.plot(ax=axes[1])
axes[1].set_title('Mean Rainfall in 2014')
plt.tight_layout()
plt.show()
```

```python
snow1 = data1['Snowf_tavg']
snow2 = data2['Snowf_tavg']
mean_s1 = snow1.mean(dim='time')
mean_s2 = snow2.mean(dim='time')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
mean_s1.plot(ax=axes[0])
axes[0].set_title('Mean Snowfall in 2000')
mean_s2.plot(ax=axes[1])
axes[1].set_title('Mean Snowfall in 2014')
plt.tight_layout()
plt.show()
```

```python
soil1 = data1['SoilMoist_P_tavg']
soil2 = data2['SoilMoist_P_tavg']
```

```python
mean_soil1 = soil1.mean(dim='time')
mean_soil2 = soil2.mean(dim='time')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
mean_soil1.plot(ax=axes[0])
axes[0].set_title('Mean Soil Moisture in 2000')
mean_soil2.plot(ax=axes[1])
axes[1].set_title('Mean Soil Moisture in 2014')
plt.tight_layout()
plt.show()
```

```python
air1 = data1['Tair_f_tavg']
air2 = data2['Tair_f_tavg']
```

## Section 3: Specific Humidity Comparison
This section compares the specific humidity for the years 2000 and 2014.

```python
h_air1 = data1['Qair_f_tavg']
h_air2 = data2['Qair_f_tavg']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
h_air1.mean(dim='time').plot(ax=axes[0])
axes[0].set_title('Specific Humidity in 2000')
h_air2.mean(dim='time').plot(ax=axes[1])
axes[1].set_title('Specific Humidity in 2014')
plt.tight_layout()
plt.show()
```

# Section 4: Runoff Calculation and Comparison
This section calculates and compares the runoff for the years 2000 and 2014.

```python
runoff1 = data1['Rainf_tavg'] - data1['Evap_tavg']
runoff2 = data2['Rainf_tavg'] - data2['Evap_tavg']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
runoff1.mean(dim='time').plot(ax=axes[0])
axes[0].set_title('Runoff in 2000')
runoff2.mean(dim='time').plot(ax=axes[1])
axes[1].set_title('Runoff in 2014')
plt.tight_layout()
plt.show()
```

## Section 5: NDVI Calculation and Comparison
This section calculates and compares the NDVI for the years 2000 and 2014.

```python
NDVI1 = (data1['SWdown_f_tavg'] - data1['LWdown_f_tavg']) / (data1['SWdown_f_tavg'] + data1['LWdown_f_tavg'])
NDVI2 = (data2['SWdown_f_tavg'] - data2['LWdown_f_tavg']) / (data2['SWdown_f_tavg'] + data2['LWdown_f_tavg'])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
NDVI1.mean(dim='time').plot(ax=axes[0])
axes[0].set_title('NDVI in 2000')
NDVI2.mean(dim='time').plot(ax=axes[1])
axes[1].set_title('NDVI in 2014')
plt.tight_layout()
plt.show()
```

## Section 6: Extreme Precipitation Events
- This section detects and compares extreme precipitation events for the years 2000 and 2014.
- The threshold for extreme events is set at **0.00002 (adjust based on criteria)**.

```python
threshold = 0.00002  # Define the threshold for extreme precipitation events

extreme_precip_events1 = data1['Rainf_tavg'] > threshold
extreme_precip_events2 = data2['Rainf_tavg'] > threshold
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
extreme_precip_events1.mean(dim='time').plot(ax=axes[0])
axes[0].set_title('Extreme Precipitation Events in 2000')
extreme_precip_events2.mean(dim='time').plot(ax=axes[1])
axes[1].set_title('Extreme Precipitation Events in 2014')
plt.tight_layout()
plt.show()
```
